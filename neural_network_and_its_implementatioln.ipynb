```python
#importing the required libraries
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
#train and test set
```

```python
#length of X_Train data
len(X_train)
```

```python
#length of y_train data
len(X_test)
```

```python
#we are scaling the array between 0 to 1
X_train=X_train/255
X_test=X_test/255
```

```python
#shape of first data in X_train
X_train[0].shape
```

```python
#oth index value data in X_train
X_train[0]
#its just 2d array 
```

```python
#now i want to flatten 2d array into 1d array for input as neural network
#our outout should be 60000,and784 not(60000,28,28)
X_train_flatten=X_train.reshape(len(X_train),28*28)
X_test_flatten=X_test.reshape(len(X_test),28*28)
```

```python
#now seeing the shape of flatten array 
X_train_flatten.shape
```

```python
X_test_flatten.shape
```

```python
#now seeing the flatten array at specific index
X_test_flatten[0]
```

```python
#sequential meaning i have stack of layer in my neural network
model=keras.Sequential([
    #dense means all the neurons here in one layer are  connected with every other neuron in second layer 
    #here 10 is output layer and 784 is input layer
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])
#The main goal of an optimizer is to find the optimal set of weights that minimize the loss function of the model
#optimizer allow you to train efficiently.
#loss
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
)
model.fit(X_train_flatten,y_train,epochs=5)
```

```python
#now we are evaluating the model
model.evaluate(X_test_flatten,y_test)
```

```python

prediction=model.predict(X_test_flatten)
```

```python
plt.matshow(X_test[0])
```

```python

prediction[0]
```

```python
np.argmax(prediction[0])
```

```python
y_predicted_labels=[np.argmax(i) for i in prediction ]
y_predicted_labels
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python

import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('predicted')
plt.ylabel('truth')
```

# ## now adding a hidden layer

```python



#sequential meaning i have stack of layer in my neural network
model=keras.Sequential([
    #dense means all the neurons here in one layer are  connected with every other neuron in second layer 
    #here 100 is a hidden layer neuron that should less than the input neuron
    #we use relu as activation function
    #first line is hidden layer
    keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])
#The main goal of an optimizer is to find the optimal set of weights that minimize the loss function of the model
#optimizer allow you to train efficiently.
#loss
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
)
model.fit(X_train_flatten,y_train,epochs=5)
```

```python
model.evaluate(X_test_flatten,y_test)
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python

import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('predicted')
plt.ylabel('truth')
```
